## Data Limits

This dataset contains anonymized search and engagement signals only.

It does not contain business decisions, content quality, user intent, or client-specific information.

The dataset supports decision-making based on observed patterns, but it cannot prove that refreshing a page will improve its future performance.

Some historical information may also be incomplete, so the results should be interpreted carefully.

## Data Limits

This dataset is anonymized and contains only observable search and engagement signals.

It cannot explain why a page performed well or poorly, nor can it capture business decisions, content quality, or user intent.

Some clients have shorter historical records than others, so comparisons should be interpreted carefully.

The dataset supports decision-making but does not establish cause-and-effect relationships.
